```python
!pip install wandb diffusers transformers accelerate -q
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os
import json
import wandb
import random
import matplotlib.pyplot as plt
from diffusers import UNet2DConditionModel
from transformers import CLIPTokenizer, CLIPTextModel

from kaggle_secrets import UserSecretsClient
try:
    user_secrets = UserSecretsClient()
    wandb_api_key = user_secrets.get_secret("wandb_api_key")
    wandb.login(key=wandb_api_key)
    anonymous = None
except:
    wandb.login(anonymous="must")
    print("Could not find W&B secret. Proceeding in anonymous mode.")

wandb.login()

DATA_DIR = "/kaggle/input/ayna-dataset/dataset"
config = {
    "LEARNING_RATE": 1e-4,
    "BATCH_SIZE": 8,
    "NUM_EPOCHS": 130,
    "IMAGE_SIZE": 128,
    "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
    "DATA_DIR": DATA_DIR,
    "PROJECT_NAME": "polygon-colorizer-diffusers-final",
    "ENTITY": wandb.api.default_entity,
    "TEXT_ENCODER_MODEL": "openai/clip-vit-base-patch32",
}
print(f"Using device: {config['DEVICE']}")


class PolygonDatasetDiffusers(Dataset):
    def __init__(self, data_dir, tokenizer, split='training', image_size=128):
        self.split_dir = os.path.join(data_dir, split)
        self.image_size = image_size
        self.tokenizer = tokenizer
        
        # Correctly join the path for the JSON file
        json_path = os.path.join(self.split_dir, 'data.json')
        with open(json_path, 'r') as f:
            self.data_map = json.load(f)
            
        print(f"[{split.upper()}] Found {len(self.data_map)} samples in {json_path}")
        
        # Transformation for the input (grayscale polygon)
        self.input_transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.Grayscale(num_output_channels=1),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5]) # Normalize to [-1, 1]
        ])
        
        # Transformation for the output (colored image)
        self.output_transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]) # Normalize to [-1, 1]
        ])

    def __len__(self):
        return len(self.data_map)

    def __getitem__(self, idx):
        item = self.data_map[idx]
        # Construct full path to input image
        input_img_path = os.path.join(self.split_dir, 'inputs', os.path.basename(item['input_polygon']))
        input_img = Image.open(input_img_path).convert("L")
        input_tensor = self.input_transform(input_img)
        
        color_name = item['colour']
        # Tokenize the color name for text conditioning
        text_inputs = self.tokenizer(
            color_name, padding="max_length", max_length=self.tokenizer.model_max_length, 
            truncation=True, return_tensors="pt"
        )
        text_input_ids = text_inputs.input_ids.squeeze(0) # Remove batch dim
        
        # Construct full path to output image
        output_img_path = os.path.join(self.split_dir, 'outputs', os.path.basename(item['output_image']))
        output_img = Image.open(output_img_path).convert("RGB")
        output_tensor = self.output_transform(output_img)
        
        return {
            "polygon_image": input_tensor,
            "text_input_ids": text_input_ids,
            "target_output": output_tensor,
            "color_name": color_name
        }

class UNetWithTanh(nn.Module):
    
    def __init__(self, unet):
        super().__init__()
        self.unet = unet
        self.tanh = nn.Tanh()

    def forward(self, sample, timestep, encoder_hidden_states, return_dict=True):
        # Pass the inputs through the original U-Net
        result = self.unet(sample=sample, timestep=timestep, encoder_hidden_states=encoder_hidden_states, return_dict=return_dict)
        
        # Apply Tanh to the sample output to scale it to [-1, 1]
        result.sample = self.tanh(result.sample)
        
        return result

tokenizer = CLIPTokenizer.from_pretrained(config['TEXT_ENCODER_MODEL'])

train_dataset = PolygonDatasetDiffusers(config['DATA_DIR'], tokenizer, split='training', image_size=config['IMAGE_SIZE'])
val_dataset = PolygonDatasetDiffusers(config['DATA_DIR'], tokenizer, split='validation', image_size=config['IMAGE_SIZE'])

train_loader = DataLoader(train_dataset, batch_size=config['BATCH_SIZE'], shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=config['BATCH_SIZE'], shuffle=False, num_workers=2)

text_encoder = CLIPTextModel.from_pretrained(config['TEXT_ENCODER_MODEL']).to(config['DEVICE'])
text_encoder.requires_grad_(False) # Freeze the text encoder

# Instantiate the base U-Net
base_unet = UNet2DConditionModel(
    in_channels=1,
    out_channels=3,
    block_out_channels=(128, 128, 256, 512),
    down_block_types=("DownBlock2D", "DownBlock2D", "CrossAttnDownBlock2D", "CrossAttnDownBlock2D"),
    up_block_types=("CrossAttnUpBlock2D", "CrossAttnUpBlock2D", "UpBlock2D", "UpBlock2D"),
    cross_attention_dim=text_encoder.config.hidden_size,
)

# Wrap the U-Net with our Tanh activation layer
unet = UNetWithTanh(base_unet).to(config['DEVICE'])


def train_one_epoch(train_loader, unet, text_encoder, optimizer, loss_fn, device):
    unet.train()
    total_loss = 0.0
    for batch in train_loader:
        polygon_images = batch['polygon_image'].to(device)
        text_input_ids = batch['text_input_ids'].to(device)
        target_outputs = batch['target_output'].to(device)
        
        with torch.no_grad():
            encoder_hidden_states = text_encoder(text_input_ids).last_hidden_state
        
        # Forward pass through our wrapped U-Net
        # We still pass a dummy timestep=0 as the model architecture requires it.
        predictions = unet(sample=polygon_images, timestep=0, encoder_hidden_states=encoder_hidden_states).sample
        loss = loss_fn(predictions, target_outputs)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        
    return total_loss / len(train_loader)

def validate_and_log_images(val_loader, unet, text_encoder, loss_fn, device, epoch):
    unet.eval()
    total_loss = 0.0
    logged_images = []
    
    with torch.no_grad():
        for i, batch in enumerate(val_loader):
            polygon_images = batch['polygon_image'].to(device)
            text_input_ids = batch['text_input_ids'].to(device)
            target_outputs = batch['target_output'].to(device)
            
            encoder_hidden_states = text_encoder(text_input_ids).last_hidden_state
            
            predictions = unet(sample=polygon_images, timestep=0, encoder_hidden_states=encoder_hidden_states).sample
            loss = loss_fn(predictions, target_outputs)
            total_loss += loss.item()
            
            if i == 0:
                # Denormalize images from [-1, 1] to [0, 1] for visualization
                # The Tanh layer in our wrapper ensures predictions are already in [-1, 1]
                preds_denorm = (predictions + 1) / 2
                targets_denorm = (target_outputs + 1) / 2
                polys_denorm = (polygon_images + 1) / 2
                
                # Convert the 1-channel input polygon to 3 channels for concatenation
                polys_denorm_rgb = polys_denorm.repeat(1, 3, 1, 1)
                
                for j in range(min(8, predictions.shape[0])):
                    color_name = batch['color_name'][j]
                    # Concatenate [Input, Ground Truth, Prediction] horizontally
                    combined_image = torch.cat([polys_denorm_rgb[j].cpu(), targets_denorm[j].cpu(), preds_denorm[j].cpu()], dim=2)
                    logged_images.append(wandb.Image(
                        combined_image,
                        caption=f"Epoch {epoch} | Color: {color_name}"
                    ))

    avg_loss = total_loss / len(val_loader)
    wandb.log({"val_loss": avg_loss, "val_predictions": logged_images})
    return avg_loss

wandb.init(project=config['PROJECT_NAME'], entity=config['ENTITY'], config=config)

loss_fn = nn.L1Loss()
optimizer = torch.optim.Adam(unet.parameters(), lr=config['LEARNING_RATE'])
best_val_loss = float('inf')

print("Starting training with wrapped Diffusers U-Net...")
for epoch in range(config['NUM_EPOCHS']):
    train_loss = train_one_epoch(train_loader, unet, text_encoder, optimizer, loss_fn, config['DEVICE'])
    val_loss = validate_and_log_images(val_loader, unet, text_encoder, loss_fn, config['DEVICE'], epoch)
    
    print(f"Epoch {epoch+1}/{config['NUM_EPOCHS']} -> Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
    wandb.log({"epoch": epoch, "train_loss": train_loss})
    
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        # Save the underlying U-Net, not the wrapper
        unet.unet.save_pretrained("best_unet_model")
        print(f"-> Saved new best model with val_loss: {best_val_loss:.4f}")

wandb.finish()
print("Training complete.")


base_unet_inference = UNet2DConditionModel.from_pretrained("./best_unet_model")
unet_inference = UNetWithTanh(base_unet_inference).to(config['DEVICE'])
unet_inference.eval()
print("Loaded best model for inference.")

def predict_diffusers(unet, text_encoder, tokenizer, image_path, color_name, device):
    image_size = config['IMAGE_SIZE']
    
    input_transform = transforms.Compose([
        transforms.Resize((image_size, image_size)),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5], std=[0.5])
    ])
    
    input_img = Image.open(image_path).convert("L")
    input_tensor = input_transform(input_img).unsqueeze(0).to(device)
    
    text_inputs = tokenizer(
        color_name, padding="max_length", max_length=tokenizer.model_max_length, 
        truncation=True, return_tensors="pt"
    )
    text_input_ids = text_inputs.input_ids.to(device)
    
    with torch.no_grad():
        text_embeddings = text_encoder(text_input_ids).last_hidden_state
        prediction = unet(sample=input_tensor, timestep=0, encoder_hidden_states=text_embeddings).sample
    
    prediction_denorm = (prediction.squeeze(0).cpu() + 1) / 2
    output_image_np = prediction_denorm.permute(1, 2, 0).numpy()
    
    return input_img, output_image_np

val_samples = [val_dataset.data_map[i] for i in random.sample(range(len(val_dataset)), 4)]

fig, axes = plt.subplots(len(val_samples), 3, figsize=(10, 12))
fig.suptitle("Model Inference Results (Fixed Diffusers UNet)", fontsize=16)

for i, sample in enumerate(val_samples):
    input_path = os.path.join(val_dataset.split_dir, 'inputs', os.path.basename(sample['input_polygon']))
    color = sample['colour']
    
    original_poly, generated_img = predict_diffusers(unet_inference, text_encoder, tokenizer, input_path, color, config['DEVICE'])
    
    gt_path = os.path.join(val_dataset.split_dir, 'outputs', os.path.basename(sample['output_image']))
    ground_truth_img = Image.open(gt_path)

    axes[i, 0].imshow(original_poly, cmap='gray')
    axes[i, 0].set_title(f"Input Polygon\nColor: '{color}'")
    axes[i, 0].axis('off')
    
    axes[i, 1].imshow(generated_img)
    axes[i, 1].set_title("Model's Output")
    axes[i, 1].axis('off')
    
    axes[i, 2].imshow(ground_truth_img)
    axes[i, 2].set_title("Ground Truth")
    axes[i, 2].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
